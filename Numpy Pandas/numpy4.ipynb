Slicing Ops

```python
x= np.arange(1,51)
x
y=x.reshape(10,5)
y
```

```python
y[0]
```

```python
y[1]
```

```python
print(y[0,0])
```

```python
print(y[2,4])
```

```python
print(y[2:5]) # last index 5 not inclusive
```

```python
print(y[0:10])
```

```python
print(y[:10])
```

```python
print(y[3:])
```

```python
print(y[:])
```

```python
print(y[:,2]) # printing only the 2nd column
```

```python
print(y[1:,3]) # from 1st row till the end but only 3rd col values
```

```python
print(y[:,2:5])
```
